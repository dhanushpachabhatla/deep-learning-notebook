#Normal epoch process vs early stopping

* early stopping reduce overfitting which increases as epochs increase usually, you can see in the below plot

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from pylab import rcParams
import matplotlib.pyplot as plt
import warnings
from mlxtend.plotting import plot_decision_regions
from matplotlib.colors import ListedColormap
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
import seaborn as sns
```

```python
X, y = make_circles(n_samples=100, noise=0.1, random_state=1)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x = X[:,0], y = X[:,1],hue=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
model = Sequential()

model.add(Dense(256, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3500, verbose=0)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

* here the optimal point is somewhere near 300, so if we apply early stopping we should get a point where our val_acc/acc is least and does not increase by future patient epochs

```python
plot_decision_regions(X_test, y_test.ravel(), clf = model, legend=2)
plt.show()
```

#Early Stopping

```python
model = Sequential()

model.add(Dense(256, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=30,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
)
```

image.png

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3500, callbacks=callback)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
plot_decision_regions(X_test, y_test.ravel(), clf = model, legend=2)
plt.show()
```
